```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Main URL of the website
base_url = 'https://agoweitu.com'  # Replace with your main URL

# Send a GET request to the base URL
response = requests.get(base_url)
response.raise_for_status()

# Parse the HTML content of the main page
soup = BeautifulSoup(response.text, 'html.parser')

# List to store navigation items and their corresponding page content
nav_items = []

# Find the navigation bar and extract links
for nav in soup.find_all('nav'):  # Adjust if needed based on actual HTML structure
    for link in nav.find_all('a', href=True):
        # Extract link text and URL
        text = link.get_text(strip=True)
        href = link['href']

        # Check if the link is relative or absolute
        if href.startswith('http'):
            full_url = href
        else:
            full_url = base_url.rstrip('/') + '/' + href.lstrip('/')  # Ensure proper URL formatting

        # Send a request to each link
        try:
            page_response = requests.get(full_url)
            page_response.raise_for_status()
            page_soup = BeautifulSoup(page_response.text, 'html.parser')

            # Initialize to collect each subheading and its text
            sub_items = []

            # Assuming subheadings are within <h2> or <h3> and content paragraphs follow
            for subheading in page_soup.find_all(['h2', 'h3']):  # Adjust based on your actual structure
                subheading_text = subheading.get_text(strip=True)
                sub_content = []

                # Collect all paragraphs under each subheading until the next heading
                for sibling in subheading.find_next_siblings():
                    if sibling.name in ['h2', 'h3']:
                        break  # Stop if the next subheading is reached
                    if sibling.name == 'p':
                        sub_content.append(sibling.get_text(strip=True))

                # Join all paragraph content under the subheading
                sub_items.append({'Subheading': subheading_text, 'Content': " ".join(sub_content)})

            # Add data to the list with navigation item, link, and structured content
            nav_items.append({
                'Navigation Item': text,
                'Link': full_url,
                'Content': sub_items
            })

        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred for {full_url}: {http_err}")
            nav_items.append({
                'Navigation Item': text,
                'Link': full_url,
                'Content': "Failed to retrieve content due to HTTP error"
            })
        except Exception as err:
            print(f"An error occurred for {full_url}: {err}")
            nav_items.append({
                'Navigation Item': text,
                'Link': full_url,
                'Content': "Failed to retrieve content due to an error"
            })

# Flatten data for DataFrame
data = []
for item in nav_items:
    if isinstance(item['Content'], list):
        for sub in item['Content']:
            data.append({
                'Navigation Item': item['Navigation Item'],
                'Link': item['Link'],
                'Subheading': sub.get('Subheading', ''),
                'Sub Content': sub.get('Content', '')
            })
    else:
        data.append({
            'Navigation Item': item['Navigation Item'],
            'Link': item['Link'],
            'Subheading': '',
            'Sub Content': item['Content']
        })

# Convert list to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('agoweitu_nav_items_and_subcontent.csv', index=False)
print("Navigation items, subheadings, and their content have been successfully saved to agoweitu_nav_items_and_subcontent.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import time

# Base URL of the website
base_url = 'https://agoweitu.com/news_item.php?item_id='

# Range of item_ids to fetch
start_id = 1
end_id = 1000  # Adjust this range based on the actual number of items

# File to save the extracted content
output_file = 'agoweitu.txt'

with open(output_file, 'w', encoding='utf-8') as file:
    for item_id in range(start_id, end_id + 1):
        url = f'{base_url}{item_id}'
        try:
            response = requests.get(url)
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')

            # Extract the main content; adjust the selector as needed
            content_div = soup.find('div', class_='content')  # Modify based on actual HTML structure
            if content_div:
                content = content_div.get_text(separator='\n', strip=True)
                # Write content to the file
                file.write(f'Item ID: {item_id}\n')
                file.write(content)
                file.write('\n' + '='*80 + '\n')
                print(f'Successfully saved content for item_id {item_id}')
            else:
                print(f'Content not found for item_id {item_id}')
        except requests.exceptions.HTTPError as http_err:
            print(f'HTTP error occurred for item_id {item_id}: {http_err}')
        except Exception as err:
            print(f'An error occurred for item_id {item_id}: {err}')
        # Respectful scraping: wait before the next request
        time.sleep(1)  # Adjust the sleep time as needed
```
